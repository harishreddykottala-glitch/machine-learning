```python
import torch
import torch.nn as nn
```

```python
model = nn.Sequential(
    nn.Linear(2, 4),
    nn.ReLU(),
    nn.Linear(4, 1)
)
```



*   X contains input features
*   y contains expected outputs



```python
X = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
y = torch.tensor([[0.],[1.],[1.],[0.]])
```

**Loss Function and Optimizer**



*   MSELoss measures prediction error
*   Adam optimizer updates weights efficiently



```python
loss_fn = nn.MSELoss()
opt = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
for _ in range(1000):
  opt.zero_grad()
  loss = loss_fn(model(X), y)
  loss.backward()
  opt.step()
```

**Model Inference**

we print the model prediction after training

```python
print(model(X))
```
